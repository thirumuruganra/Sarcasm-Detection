```python
import pandas as pd
from googletrans import Translator

translator = Translator()

def translate_text(text, i):
    try:
        translation = translator.translate(text, src='ta', dest='en')
        print(i, translation.text)
        return translation.text
    except Exception as e:
        print(f"Error translating text at index {i}: {text}\n{e}")
        return text

batch_size = 100
num_batches = len(df) // batch_size + (1 if len(df) % batch_size != 0 else 0)

# Track completed batches
completed_batches = 0
checkpoint_file = '/content/drive/My Drive/Sarcasm Detection/checkpoint1.txt'

# Check if there's an existing checkpoint
try:
    with open(checkpoint_file, 'r') as f:
        completed_batches = int(f.read().strip())
except FileNotFoundError:
    pass

for i in range(completed_batches, num_batches):
    start_idx = i * batch_size
    end_idx = min((i + 1) * batch_size, len(df))

    batch = df.iloc[start_idx:end_idx].copy()
    batch['Translated_Text'] = batch.apply(lambda row: translate_text(row['Text'], row.name), axis=1)

    # Append the translated batch to the output CSV file
    if i == 0 and completed_batches == 0:
        batch.to_csv('/content/drive/My Drive/Sarcasm Detection/Malayalam/translated_output_dev.csv', index=False)
    else:
        batch.to_csv('/content/drive/My Drive/Sarcasm Detection/Malayalam/translated_output_dev.csv', mode='a', header=False, index=False)

    # Update checkpoint
    with open(checkpoint_file, 'w') as f:
        f.write(str(i + 1))

    print(f"Batch {i+1}/{num_batches} translated and saved.")

print("Translation completed and saved to 'translated_output_dev.csv'")

# Remove checkpoint file after completion
import os
os.remove(checkpoint_file)
```
